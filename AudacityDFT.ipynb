```python
from visual import *
from visual.graph import *
import cmath

data = open('sample-data.txt', 'r')
data1 = data.readlines()
plot = gcurve(color = color.green)
plot1 = gcurve(color = color.yellow)
plot2 = gcurve(color = color.cyan)
i = 0
fourierData = []
for line in data1:
    rate(1000)
    i = i + 1
    output = line.split()
    if i > 6:
        plot.plot(pos = (i-7, float(output[0])))
        fourierData.append(float(output[0]))
        
peakF = 0
peakC = 0
F = []
Amp = []
Phase = []

magZPrev2 = 0
magZPrev = 0
magZ = 0
        
for n in range(len(fourierData)/6):
    rate(100000000)
    totalSumR = 0 #real component of the sum
    totalSumC = 0 #complex component of the sum
    magZPrev2 = magZPrev
    magZPrev = magZ
    for z in range(len(fourierData)):
        rate(100000000)
        totalSumR = totalSumR + fourierData[z]*cos(-2*pi*n*float(z)/len(fourierData))
        totalSumC = totalSumC + fourierData[z]*sin(-2*pi*n*float(z)/len(fourierData))
    magZ = sqrt(totalSumR**2 + totalSumC**2)
    F.append(n/(len(fourierData)/44100.0))
    Amp.append(magZ)
    Phase.append(atan(totalSumC/totalSumR))
    plot1.plot(pos = (n/(len(fourierData)/44100.0), magZ))
    if magZ > peakC:
        peakF = n/(len(fourierData)/44100.0)
        peakC = magZ
    if magZPrev > magZ and magZPrev > magZPrev2:
        print(str((n-1)/(len(fourierData)/44100.0)) + " Hz, Amplitude = " + str(magZPrev) + ", Phase = " + 
             str(atan(totalSumC/totalSumR)))
print("Peak Frequency = " + str(peakF) + " Hz")
while true:
    rate(100)
    true
```
